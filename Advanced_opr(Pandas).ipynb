```python
# To create a new column and store : if Fare<10 euro : print: cheap,
# if 10<Fare<20 : print :Mid, 
# if Fare>20 : print : High
```

```python
def cat_fare(x):
    if x<10:
        return "cheap"
    elif x>=10 and x<20:
        return "Mid"
    else:
        return "high"
```

```python
df['travel_Fare'] = df['Fare'].apply(cat_fare)
```
